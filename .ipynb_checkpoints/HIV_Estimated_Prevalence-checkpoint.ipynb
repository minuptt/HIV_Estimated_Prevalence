 # Import the libraries

```python
import pandas as pd
import pycountry
import pycountry_convert as pc
import numpy as np
import matplotlib.pyplot as plt
```

# Load data

```python
# Read and store content of an excel file to a Dataframe
org_file = pd.read_excel("/Users/thiphan/Desktop/HIV_Estimated_Prevalence/indicator hiv estimated prevalence% 15-49.xlsx")
org_file
```

# Create a new CSV file that include the "Contients" to label each country/region to an appropriate continent

```python
# create a country conversion function
# convert country to country code and from country code to continent
# 

# Perform slicing and cout the total number of rows by using count function
total_rows = org_file.iloc[:,0].count()

# Assigned all pycountry.countries data to "data" variable
data = pycountry.countries

# Appending index values
inp =  []

# Appending index values that have continents
inp_cont = []

# Appending continents
cont = []

# Appending countries
ccode = []

# Find difference between 2 lists
def diff(li1, li2):
    li_diff = [i for i in (li1 + li2) 
                       if i not in li1 
                       or i not in li2]
    return li_diff

# Appending all index values to "inp_cont" list
for i in range(total_rows):
    inp_cont.append(i)
    for a in data:
        b = str(a) #convert each element to string for comparision
        rem = b.split("numeric")[0] #split number and store at the first value in list
        temp = org_file['Estimated HIV Prevalence% - (Ages 15-49)'][i]
        try:
            if temp in rem:
                if temp not in ccode:
                    #append all countries that satisfied
                    ccode.append(temp) 
                    temp1 = temp
                    #convert country name to country code
                    country_code = pc.country_name_to_country_alpha2(temp1,cn_name_format = "default")
                    #convert country code to continent
                    continent_name = pc.country_alpha2_to_continent_code(country_code)
                    
                    #append continents to "cont" list
                    cont.append(continent_name)
                    
                    #append all countries that have continents
                    inp.append(i)
        except:
            continue
list_countries = diff(inp,inp_cont)

#find the length of countries that have no continent
length = len(list_countries)
for i in range (length):
    ab = list_countries[i]
    
    #insert no continent index as the name "No continent"
    cont.insert(ab,'No continent')

#insert new column "Continents" in dataset
org_file.insert(1,'Continents', cont, True)

#create new CSV file with updated dataset
org_file.to_csv('Updated.csv')
```

```python
#read CSV file (Updated.csv) and convert into dataframe object
df = pd.DataFrame(pd.read_csv("/Users/thiphan/Desktop/HIV_Estimated_Prevalence/Updated.csv"))

#show the dataframe
#df
```

# Highest and Lowest Average HIV estimated prevalence

```python
# Replace all null values to NA
df['Continents'] = df['Continents'].replace(np.nan, 'NA')


# Calculate mean to all of countries
mean = df[['2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011']]\
        .mean(axis = 1)
df['mean'] = mean
```

Highest Average HIV

```python
maximum = df.groupby(['Continents'], as_index = False, sort = False)['mean', 'Estimated HIV Prevalence% - (Ages 15-49)'].max()
maximum
```

Lowest Average HIV

```python
minimum = df.groupby(['Continents'], as_index = False, sort = False)['mean', 'Estimated HIV Prevalence% - (Ages 15-49)'].min()
minimum
```

```python
#create new dataframe
df_new = df[['Continents', 'mean', 'Estimated HIV Prevalence% - (Ages 15-49)']]
#df_new

df_max = pd.concat([df_new,maximum], ignore_index = True, sort = True).drop_duplicates('Estimated HIV Prevalence% - (Ages 15-49)', keep = False)
df_max_min = pd.concat([df_max, minimum], ignore_index = True, sort = True).drop_duplicates('Estimated HIV Prevalence% - (Ages 15-49)', keep = False)
#df_max_min

# select one random countries
one_rd = lambda obj: obj.loc[np.random.choice(obj.index, 1, replace = True),:]

#create new dataframe with group by continent
df_final = df_max_min.groupby('Continents', as_index = False).apply(one_rd)
df_final
```

# Create bar charts, overlaid bar chart and overlaid line chart

```python
# create figure to plot graphs
fig = plt.figure(figsize=(15,7))
ax1=fig.add_subplot(2,2,1) 
ax2=fig.add_subplot(2,2,2) 
ax3=fig.add_subplot(2,2,3) 
ax4=fig.add_subplot(2,2,4) 

# bar chart
ax1.bar(maximum['Continents'],maximum['mean'],color=('r'),label='Maximum') 
ax2.bar(minimum['Continents'],minimum['mean'],label='Minimum') 
ax3.bar(maximum['Continents'],maximum['mean']+2,width=0.4,color='r',label='Maximum')
ax3.bar(minimum.index,minimum['mean']+2,width=0.4,color='b',label='Minimum')

# overlaid line chart
ax4.plot(maximum['Continents'],maximum['mean'],color='r',label='Maximum') 
ax4.plot(minimum['Continents'],minimum['mean'],color='b',label='Minimum') 
ax4.plot(df_final['Continents'],df_final['mean'],color='y',label='Random Point')

# set title and legend
ax1.set_title("Highest Average")
ax1.set_xlabel("Continents")
ax1.set_ylabel("Maximum Mean")
ax1.legend()

ax2.set_title("Lowest Average")
ax2.set_xlabel("Continents")
ax2.set_ylabel("Minimum Mean")
ax2.legend()

ax3.set_title("Min and Max")
ax3.set_xlabel("Continents")
ax3.set_ylabel("Highest and Lowest Average")
ax3.legend()

ax4.set_title("Min-Max Avg & Rand Point")
ax4.set_xlabel("Continents")
ax4.set_ylabel("Min-Max Avg & Rand Point ")
ax4.legend()

# fit plots within figure
plt.tight_layout()
```

# Calculate the average HIV estimated prevalence and create the line chart for each continent and overlaid line chart for all continents to show their chahges of the average HIV estimated prevalence

```python
# create new dataframe by group continents and transposes x-axis and y-axis
continent_by_country = df.groupby(['Continents'], sort = False).mean().transpose()
continent = continent_by_country.drop(['Unnamed: 0'])
continent
```

```python
# create figure to plot graphs
fig=plt.figure(figsize=(20,18))

# Line chart for AS continent
ax1=fig.add_subplot(4,2,1)
ax1.plot(continent.index,continent['AS'],color='r')
ax1.set_xlabel("Years")
ax1.set_ylabel('Avg')
ax1.set_title(" AS Continent")

# Line chart for EU continent
ax2=fig.add_subplot(4,2,2)
ax2.plot(continent.index,continent['EU'],color='g')
ax2.set_xlabel("Years")
ax2.set_ylabel('Avg')
ax2.set_title(" EU Continent")

# Line chart for AF continent
ax3=fig.add_subplot(4,2,3)
ax3.plot(continent.index,continent['AF'],color='b')
ax3.set_xlabel("Years")
ax3.set_ylabel('Avg')
ax3.set_title(" AF Continent")

# Line chart for OC continent
ax4=fig.add_subplot(4,2,4)
ax4.plot(continent.index,continent['OC'],color='k')
ax4.set_xlabel("Years")
ax4.set_ylabel('Avg')
ax4.set_title(" OC Continent")

ax5=fig.add_subplot(4,2,5)
ax5.plot(continent.index,continent['NA'],color='c')
ax5.set_xlabel("Years")
ax5.set_ylabel('Avg')
ax5.set_title(" NA Continent")

# Line chart for SA continent
ax6=fig.add_subplot(4,2,6)
ax6.plot(continent.index,continent['SA'],color='m')
ax6.set_xlabel("Years")
ax6.set_ylabel('Avg')
ax6.set_title(" SA Continent")

# Overlaid line chart for all continents
ax7=fig.add_subplot(4,2,7)
ax7.plot(continent.index,continent['AS'],color='r',label='AS')
ax7.plot(continent.index,continent['EU'],color='g',label='EU')
ax7.plot(continent.index,continent['AF'],color='b',label='AF')
ax7.plot(continent.index,continent['OC'],color='k',label='OC')
ax7.plot(continent.index,continent['NA'],color='c',label='NA')
ax7.plot(continent.index,continent['SA'],color='m',label='SA')
ax7.set_xlabel("Years")
ax7.set_ylabel('Avg')
ax7.set_title(" All Continents")
ax7.legend()

# fit plots within figure
plt.tight_layout()
```
